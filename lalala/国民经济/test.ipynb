```python
import numpy as np
import matplotlib.pyplot as plt

data=np.load('国民经济核算季度数据.npz',allow_pickle=True)
print(data.files)
print(data['columns'])
print(data['values'])
```

```python
plt.figure(figsize=(10,8))

l=['-','--','-.']
c=['r','b','y']
m=['o','*','D']

plt.rcParams['font.family']='STSong'
plt.rcParams['font.size']=14
plt.xticks(range(0,70,4),data['values'][range(0,70,4),1],rotation=45)
for j,i in enumerate([3,4,5]):
    plt.plot(data['values'][:,1],data['values'][:,i],linestyle=l[j],color=c[j],marker=m[j])
plt.xlabel('季度')
plt.ylabel('生产总值（亿元）')
plt.ylim((0,115000))
plt.title('2000-2017年各季度各产业生产总值折线图')
plt.legend(['第一产业','第二产业','第三产业'])
plt.savefig('2000-2017年各季度各产业生产总值折线图.png')
plt.show()
```

```python
plt.figure(figsize=(10,8))

plt.rcParams['font.family']='STSong'
plt.rcParams['font.size']=14
plt.xticks(range(0,70,4),data['values'][range(0,70,4),1],rotation=45)
for i in [3,4,5]:
    plt.scatter(data['values'][:,1],data['values'][:,i])
plt.xlabel('季度')
plt.ylabel('生产总值（亿元）')
plt.ylim((0,115000))
plt.title('2000-2017年各季度各产业生产总值散点图')
plt.legend(['第一产业','第二产业','第三产业'])
plt.savefig('2000-2017年各季度各产业生产总值散点图.png')
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.pie(data['values'][-1,3:6],autopct='%.2f %%',explode=[0.1,0,0],labels=['第一产业','第二产业','第三产业'])
plt.title('2000-2017年各季度各产业生产总值饼图')
plt.savefig('2000-2017年各季度各产业生产总值饼图')
plt.show()
```

```python
plt.figure(figsize=(10,8))
num=(list(data['values'][:,3]),list(data['values'][:,4]),list(data['values'][:,5]))
plt.boxplot(num)
plt.title('2000-2017年各产业生产总值箱线图')
plt.savefig('2000-2017年各产业生产总值箱线图')
plt.show()
```
