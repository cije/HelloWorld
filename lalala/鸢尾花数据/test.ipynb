```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
d=np.load('iris.npz',allow_pickle=True)
print(d.files)
data=d['data']
print(data)
features_name=d['features_name']
print(features_name)
```

```python
plt.figure(figsize=(22,17))
plt.rcParams['font.family']='STSong'
plt.rcParams['font.size']=14
# m=1
for i in range(0,4):
    for j in range(0,4):
        # plt.subplot(4,4,m)
        # m=m+1
        plt.subplot(4,4,i * 4 + j + 1)
        plt.scatter(data[:,i],data[:,j])
        plt.xlabel(features_name[i])
        plt.ylabel(features_name[j])
plt.suptitle('Iris')
plt.savefig('Iris.png')
plt.show()
```

```python
plt.figure(figsize=(10,8))
num=(list(data[:,0]),list(data[:,1]),list(data[:,2]),list(data[:,3]))
plt.boxplot(num,labels=list(features_name[:-1]))
plt.title('Iris')
plt.savefig('Iris.png')
plt.show()
```
